### Question 1: Basic GroupBy Operation
Find the average salary for each department.


```python
df.groupby('Department')['Salary'].mean().reset_index()
```

```python
df.groupby('Department')['Salary'].max().reset_index()
```

### Question 2: Multi-Column Aggregation
Calculate the total bonus and the average salary for each job level in each department.

```python
df.head(3)
```

```python
df.groupby(['Department','Job_Level']).agg({'Bonus':'sum','Salary':'mean'})
```

### Question 3: Filtering and Grouping
Find the average salary of employees in each department who have more than 5 years of experience.

```python
df[df['Years_of_Experience'] > 5].groupby('Department')['Salary'].mean().reset_index()
```

### Question 4: Date-Based Grouping
Group the data by the year of joining and calculate the total bonus paid to employees each year.

```python
df.head(3)
```

```python
df.groupby(df['Join_Date'].dt.year)['Bonus'].sum()
```

```python
df['Join_Date'].dt.year
```

### Question 5: Conditional Aggregation
Calculate the average salary and performance score of employees aged above 30 for each department.

```python
df[df['Age']>30].groupby('Department').agg({'Salary':'mean','Performance_Score':'mean'})
```

### Question 6: Custom Aggregation with Multiple Functions
For each gender and job level, compute the following statistics for salaries: mean, median, and standard deviation.

```python
df.groupby(['Gender', 'Job_Level'])['Salary'].agg(['mean', 'median', 'std'])
```

### Question 7: Grouping with Multiple Conditions
Find the highest salary for each combination of department and performance score.

```python
df.groupby(['Department','Performance_Score'])['Salary'].max()
```

### Question 8: Group By and Apply Custom Functions
Create a custom function to calculate the range (max - min) of salaries for each location and apply it using groupby().

```python
def salary_range(x):
    return x.max() - x.min()

salary_range_by_location = df.groupby('Location')['Salary'].apply(salary_range)
salary_range_by_location
```

### Question 9: Counting Unique Values
Count the number of unique job levels in each department.

```python
df.groupby('Department')['Job_Level'].nunique()
```

### Question 10: Combining GroupBy with Other Operations
Find the top 3 employees with the highest bonuses in each department.

```python
df.head(5)
```

```python
df.groupby('Department').apply(lambda x: x.nlargest(3, 'Bonus'))
```

```python
df.groupby('Department')['Bonus'].nlargest(3)
```
